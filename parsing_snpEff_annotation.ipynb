```python
import numpy as np
import pandas as pd
```

```python
file_path = "/Users/tennisnyjmac/Desktop/BI/bioinformatics_institute/BI_practicum/AntibioticResistance_HW1/VarScan_results_annotated.vcf"
```

```python
parsed_data = []

with open(file_path, "r") as file:
    for line in file:
        if not line.startswith("#"):
            line = line.split('\t')
            chromosome = line[0]
            position = line[1]
            ref_allele = line[3]
            alt_allele = line[4]
            info_field = line[7]
            format_field = line[8]
            sample_field = line[9]

            ann_field = [field for field in info_field.split(';') if field.startswith('ANN=')]
            if ann_field:
                annotation = ann_field[0].replace('ANN=', '').split(',')[0]
                print(annotation)
                ann_split = annotation.split('|')
                parsed_data.append({
                    'Chromosome': chromosome,
                    'Position': position,
                    'Reference Allele': ref_allele,
                    'Alternative Allele': alt_allele,
                    'Impact': ann_split[2],  # Example: MODERATE, LOW, etc.
                    'Gene Name': ann_split[3],
                    'Gene ID': ann_split[4],
                    'Transcript ID': ann_split[6],
                    'Variant Type': ann_split[1],  # Example: missense_variant, upstream_gene_variant, etc.
                    'Amino Acid Change': ann_split[10],
                    'cDNA Position': ann_split[9],
                    'Protein Position': ann_split[11],
                    'Genotype': sample_field.split(':')[0],  # Extract genotype from FORMAT
                    'Depth': sample_field.split(':')[3]  # Extract depth (DP) from FORMAT
                })
            
df = pd.DataFrame(parsed_data)
```

```python
df.to_csv("SNV_annotation_snpEff.csv")
```
